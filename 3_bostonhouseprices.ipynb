# Load Boston House Prices Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
```

```python
dataBoston = load_boston()
print(dataBoston)
```

```python
print(dir(dataBoston))
```

```python
print(dataBoston['filename'])
```

### Nama Kolom (feature_names)

```python
print(dataBoston['feature_names'])
```

```python
print(dataBoston['data'])
```

```python
print(dataBoston['data'].shape)
```

### Dependent (target)

```python
print(dataBoston['target'])
```

```python
df = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
df
```

```python
df['price'] = dataBoston['target']
df
```

### Correlation Check

```python
corr = df.corr()
corr
```

```python
plt.imshow(corr, cmap='BuPu')
plt.colorbar()
plt.xticks(np.arange(14), list(df.columns), rotation=90)
plt.yticks(np.arange(14), list(df.columns))
plt.show
```

### Model Machine Learning Regression

```python
model = LinearRegression()
model.fit(df[['ZN', 'CHAS', 'RM', 'DIS', 'B']], df['price'])
```

### Predict

```python
model.predict(df.head(1)[['ZN', 'CHAS', 'RM', 'DIS', 'B']])
```
